```
import pandas as pd
import numpy as np
```

```
phone = pd.read_csv('https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/cell_phones.csv')
pop =  pd.read_csv('https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/population1.csv')
cou_codes = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--entities--geo--country.csv')
```

```
cou_codes.head()
```

```
merged_pho_pop = pd.merge(pop, phone, on = ('Country','year'))
merged_pho_pop.shape
```

```
cou_codes.head(5)
```

```
cou_codes.rename(columns = {'country':'geo', 'name':'Country'}, inplace = True)
```

```
cou_codes.head(5)
```

```
geo_cou = cou_codes[['geo','Country']]
# geo_cou.head()
```

```
dataset = pd.merge(merged_pho_pop, geo_cou)
dataset.head(50)
```

```
dataset['cell_per'] = dataset['cell_phones_total'] / dataset['population']
```

```
dataset['geo'] = dataset['geo'].str.upper()
dataset.head(5)
```
